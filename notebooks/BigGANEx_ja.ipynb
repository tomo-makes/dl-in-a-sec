# 画像生成

画像生成は、ランダムな入力データから、リアルな画像を生成することを指します。通常、[Ian Goodfellow et al](http://papers.nips.cc/paper/5423-generative-adversarial-nets)が2014年に発表した、敵対的生成ネットワーク（GAN）を使って実現します。 GANは通常、2つのネットワークを持ちます。

- 最初のネットワークは、偽の画像と実際の画像を区別できるDiscriminatorです
- もうひとつのネットワークは、ランダムデータを与えられたら、学習データの分布を学習し、リアルな画像を生成できるGeneratorです

GANは2014年以来、進化してきています。

![alt text](https://pbs.twimg.com/media/Deex0DoU0AABQK_.jpg)

たとえば、ImageNetのような大規模データセットで学習を進める時、問題になるのは学習プロセスの安定性です。学習プロセスが不安定だと、望むようなリアルな画像生成は望めません。通常、生成画像の品質と多様性を評価する測定基準にインセプションスコア **IS** を使います。ここでは、生成画像の確率分布からサンプリングされた$ x \sim p_g $を考えます。


$$\mathrm{IS}(G) = \mathrm{exp}\left(\mathbb{E}_{x\sim p_g} D_{KL}  (p(y|x) || p(y))\right)$$

生成画像の統計的評価のために、Inception v3モデルを使います。 $ KL $は、KLダイバージェンスを指します。これは、2つの分布の「近さ」を評価するための尺度です。


もうひとつの測定基準は**FID** (Fréchet Inception Distance)です。現実の画像$ x $と生成画像$ g $が与えられると、スコアを次のように計算します。

$$\mathrm{FID} (x,g) = \Vert \mu_x - \mu_g\Vert_2^2 + \mathrm{Tr}\left( \Sigma_x+\Sigma_g  - 2 \sqrt{\Sigma_x \Sigma_g}\right) $$


ここで、$ \mathrm {Tr} $は行列の対角和（トレース）で、$ \Sigma $は共分散行列です。

# BigGAN

18年9月下旬に、DeepMindの [Andrew Brock et al.](https://arxiv.org/abs/1809.11096) による "Large Scale GAN Training for High Fidelity Natural Image Synthesis" （高精細自然画像生成のための大規模GAN学習）が発表されました。画像生成の最先端の成果といえるでしょう。以前の最高であったIS=52.52およびFID=18.65を大きく上回り、IS=166.3およびFID=9.6をマークしました。以下の画像がコンピュータで生成されたとは、にわかに信じがたいでしょう。

![alt text](https://cdn-images-1.medium.com/max/2000/1*Yw2KxjmIkj8yqS-ykLCQCQ.png)

BigGANは、より大きなネットワークおよびバッチサイズを使い、以前の方法を拡張しました。論文によると、

「 GANはスケーリングから劇的に恩恵を受けて、従来技術と比較し2〜4倍のパラメータ、8倍のバッチサイズでモデルを学習させます。」

とされています。モデルはGoogle TPUの128から512コアという莫大な計算資源で学習しました。BigGANは、分類クラスを条件として与えられるようGeneratorを学習するために、分類クラスのラベルが埋め込みレイヤに挿入される、条件付きGANです。さらに著者らは、潜在ベクトルが異なるレベルの特徴に作用できるように、階層的な潜在空間の変形を使いました。つまり、ノイズベクトルzが、Generatorの複数のレイヤに挿入されるのです。

# 必要なライブラリのインポート

```
from IPython import display
import numpy as np
from scipy.stats import truncnorm
import tensorflow as tf
import tensorflow_hub as hub
import matplotlib.pyplot as plt
from numpy import linalg as LA
```

# TF HubからBigGANのモデルを取得する

[TensorFlow Hub](https://www.tensorflow.org/hub/)を使えば、学習済みの機械学習モジュールを再利用し、実験できます。各モジュールは自己完結型のTensorFlowグラフです。 BigGANは、3つの異なるサイズ128、256、512で使えます。今回の実行で使うため、いずれか1つを選択してください。

```
tf.reset_default_graph()
model_size = "3)biggan-512" #@param ["1)biggan-128" , "2)biggan-256" , "3)biggan-512"]
which_model = model_size.split(')')[1]
module_path = 'https://tfhub.dev/deepmind/'+which_model+'/2'
module = hub.Module(module_path)
```

モデルの入出力を確認します。

```
inputs = {k: tf.placeholder(v.dtype, v.get_shape().as_list(), k)
          for k, v in module.get_input_info_dict().items()}
output = module(inputs)

print ('Inputs:\n', '\n'.join(
        '{}: {}'.format(*kv) for kv in inputs.items()))

print ('Output:', output)
```

サンプリングするラベルの数と潜在空間のサイズ$ z $を格納します。


```
vocab_size = inputs['y'].get_shape().as_list()[1]
latent_size = inputs['z'].get_shape().as_list()[1]

print('Number of labels ', vocab_size)
print('The size of the latent space ', latent_size)
```

このモデルは、ラベル入力を1つのホットエンコードとして受け取ります。各クラス $c \in [0, 1000)$ を、対応するクラスの要素を除き全て0で埋めた、サイズ$ 1000 $のベクトルにマッピングします。

```
#helper functions to generate the encoding
def one_hot(index, vocab_size=1000):
  index = np.asarray(index)
  if len(index.shape) == 0:
    index = np.asarray([index])
  assert len(index.shape) == 1
  num = index.shape[0]
  output = np.zeros((num, vocab_size), dtype=np.float32)
  output[np.arange(num), index] = 1
  return output
```

# 切り捨てのトリック

GANに関するこれまでの研究では、潜在ベクトル　$z \sim \mathcal{N}(0, I)$を恒等共分散行列を使った通常の分布としてサンプリングしています。 BigGANの作者であるOTOHは、$a \in \mathbb{R}^+$に特定の領域$[-a, a]$で切り捨てた通常の分布を使いました。これにより、**IS**と**FID**の両方のスコアが向上しました。しかし欠点は、ベクトルサンプリングの全体的な種類が減少することです。したがって、Generator **G** のサンプルの品質と多様性の間はトレードオフの関係です。

ここでは、デフォルトの切り捨て閾値を$ 2 $に設定します。つまり、$ z $値は、領域$ [ -  2、2] $からサンプリングされます。オプションの変数 `seed`は整数値を取り、それは再サンプリング時に、同じ擬似乱数が生成されるように使います。したがって、同じ `seed` を使用すると、同じ$ z $ベクトルが得られます。

```
def get_zy(index, trunc = 1., batch_size = 1, seed = None):
  #convert the label to one hot encoding 
  y = one_hot(index)
  
  #sample a batch of z-vectors 
  z = truncnorm.rvs(-2, 2, size=(batch_size, latent_size), random_state = np.random.RandomState(seed)) * trunc
  return z, y
```

正規分布の場合、ゼロに近い値が高い確率で選択されることに注意してください。

```
z, _ = get_zy(np.random.choice(range(0, vocab_size), size = 1),  seed = 2)

y = np.linspace(0, latent_size, latent_size)

plt.scatter(y, z)
plt.show()
```

# モデルをテストする


inputのセッションとともに、グラフを動作させます。そして、入力を与えます。

```
def generate(sess, z, y, trunc = 1.):
  feed_dict = {inputs['z']: z, inputs['y']:y, inputs['truncation']: trunc}
  im = sess.run(output, feed_dict=feed_dict)
  
  #postprocess the image 
  im = np.clip(((im + 1) / 2.0) * 256, 0, 255)
  im = np.uint8(im)  
  im = im.squeeze()
  return im
```

# 潜在空間の補間

2つのベクトル$v_1、v_2$と整数$N$が与えられたとき、それらの間を補間したい、すなわち2つのベクトル間を$ N $ベクトルで移動するマッピングを作成します。

## 線形補間

単純な関数を使うのが最も単純なタイプの補間です

$$F(v_1, v_2, N) = x v_2 + (1-x) v_1 \,\, ,\,\,\, x \in \left(0,\frac{1}{N}, \cdots, \frac{N}{N}\right)$$

```
def interpolate_linear(v1, v2, num_steps):
  vectors = []
  for x in np.linspace(0.0, 1.0, num_steps):
    vectors.append(v2*x+v1*(1-x))  
  return np.array(vectors)
```

## 超球面補間

反復ごとに$i \in (0, N-1)$

$$ z = v_1 + \left( v_2 \frac{\Vert v_1 \Vert }{\Vert v_2 \Vert}  - v_1\right) \times \frac{i}{N-1}$$


それから$ z $を以下のように更新します。

$$\hat{z} =  \Vert v_1 \Vert \times \frac{z}{\Vert z \Vert}$$

```
def interpolate_hypersphere(v1, v2, num_steps):

  v1_norm = LA.norm(v1)
  v2_norm = LA.norm(v2)
  v2_normalized = v2 * (v1_norm / v2_norm)

  vectors = []
  for step in range(num_steps):
    interpolated = v1 + (v2_normalized - v1) * step / (num_steps - 1)
    interpolated_norm =  LA.norm(interpolated)
    interpolated_normalized = interpolated * (v1_norm / interpolated_norm)
    vectors.append(interpolated_normalized)
  return np.array(vectors)
```

グラフからわかるように、正弦関数はゼロに近い値を保持しますが、値を0.5から離れるように変化させるため、背景が変わります。

```
with tf.Session() as sess:
  images = None
  #initialize the variables
  sess.run(tf.global_variables_initializer())
  
  #truncation value
  trunc = 1.0
  
  #choose a class 
  index = 207
  
  z, y = get_zy(index, trunc = trunc)
  im1 = generate(sess, z, y, trunc = trunc)
  im2 = generate(sess, np.sin(z), y, trunc = trunc)

  images = np.concatenate([im1, im2], axis = 1)
  #show the image 
  plt.axis('off')
  plt.imshow(images)
  plt.show()
```

## 実験 2

生成された画像の詳細を確認するために、特定のカテゴリを詳しく見ていきます。そのためにはベクトルの重みを増やします。ベクトル空間の各特徴が1か-1の値でなければ機能しません。そのため、ベクトルの各値をその絶対値、すなわち$ z / | z | $で除算します。それから負の値を増やして乗算し、スケーリングをします。これは選択する `seed` に大きく依存しますので注意してください。ここでは **seed = 2** を選びます。

```
with tf.Session() as sess:
  images = None
  #initialize the variables
  sess.run(tf.global_variables_initializer())
  
  #truncation value 
  trunc = 1.0
  
  #choose a class 
  index = 300
  
  #sample a vector with the same seed 
  z, y = get_zy(index, seed = 2)
  
  #generate the images with different scaling factor 
  im1 = generate(sess, -0.5*(z/np.abs(z)), y)
  im2 = generate(sess, -1.0*(z/np.abs(z)), y)
  im3 = generate(sess, -1.5*(z/np.abs(z)), y)


  images = np.concatenate([im1, im2, im3], axis = 1)
  #show the image 
  plt.axis('off')
  plt.imshow(images)
  plt.show()
```

`seed` を **seed = 8** に変えます。これは色付けプロセスを適用します。

```
with tf.Session() as sess:
  images = None
  #initialize the variables
  sess.run(tf.global_variables_initializer())
  
  #truncation value 
  trunc = 1.0
  
  index = 200
 
        
  #sample a vector with the same seed 
  z, y = get_zy(index, seed = 8)

  #generate the images with different scaling factor 
  im1 = generate(sess, -1.0*(z/np.abs(z)), y)
  im2 = generate(sess, -1.5*(z/np.abs(z)), y)


  images = np.concatenate([im2, im1], axis = 1)
  #show the image 
  plt.axis('off')
  plt.imshow(images)
  plt.show()
```

## 実験 3

この実験では、2つの異なるクラス間を行き来します。考え方は単純で、インコードされたラベルを平均して同じシードベクトルを使用するだけです。以下の数式で組み合わせます:

$$ \hat {y} = ay_1 +（1-a）y_2 $$

$ a = 0.5 $を使用すると、2つのカテゴリが結合されることに注意してください。 $ a <0.5 $の場合、生成された画像は$ y_2 $に近くなり、$ a> 0.5 $の場合は$ y_1 $に近くなります。

```
with tf.Session() as sess:
  images = None
  #initialize the variables
  sess.run(tf.global_variables_initializer())
  
  #truncation value 
  trunc = 1.0
  
  i1 = 207
  i2 = 288
  
  #closeness variable 
  a = 0.5
        
  #sample a vector with the same seed 
  z, y1 = get_zy(i1)
  z, y2 = get_zy(i2)
  
  #generate the images with different scaling factor 
  im1 = generate(sess, z, y1)
  im2 = generate(sess, z, y2)
  im3 = generate(sess, z, a*y1+(1-a)*y2)

  images = np.concatenate([im1, im2, im3], axis = 1)
  #show the image 
  plt.axis('off')
  plt.imshow(images)
  plt.show()
```
